# 3. 
AND, OR, NOT Boolean operators Truth tables

```python

#or-truth table

print(1 or 0)         # 0-False
print(1 or 1)         # 1-True
print(1 or 0)
print(0 or 0)


#and-truth table

print(1 and 0)
print(0 and 1)
print(1 and 1)
print(0 and 0)


#not-truth table

print(not 1) #not true means false(0)
print(not 0) #not false means true(1)
```

# 11.
There is no difference between range(0), range(0,10) , range(0,10,1).
'range(10)' means- intergers between 0 to 9 . This is more clearly quoted as 'range(0,10)'.
And 'range(0,10,1)' means begin from 0 till 10 and 10 not included, 1 illustrates to be move in positive direction .

By default 'range(10)' means nothing but move in positive direction from the default value 0 to 9.



```python
for i in range(10):
    print(i)
```

```python
for i in range(0,10):
    print(i)
```

```python
for i in range(0,10,1):
    print(i)
```
